```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import re
import string

def text_temizle1(text):
    #kucuk harf yap
    text = text.lower()
    #parantez içindeki yazıları sil
    text = re.sub('\[.*?\]', '', text)
    #noktalama işaretlerini sil
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    #sayıları sil
    text = re.sub('\w*\d\w*', '', text)
    #3 karakterden kısa olan sözcükleri sil türkçe karakterlerle birlikte
    text = re.sub(r'\b[a-zğüşöçıİĞÜŞÖÇ]{1,2}\b', r'',text)
    return text
```

```python
dataFrame = pd.read_csv("film_Series.csv")
dataFrame.head(3)
```

```python
columns = ['Name','Votes','Rate','Duration','Genre','Type','Nudity','Violence','Profanity','Alcohol','Frightening']
```

```python
dataFrame[columns].head(3)
```

```python
dataFrame.isnull().values.any()
```

```python
def get_important_features(data):
    important_features = []
    for i in range(0, data.shape[0]):
        important_features.append(data['Name'][i] + ' ' + data['Votes'][i] + ' ' + data['Rate'][i] + ' ' + data['Duration'][i] + ' ' + data['Genre'][i] + ' ' + data['Type'][i] + ' ' + data['Nudity'][i] + ' ' + data['Violence'][i] + ' ' + data['Profanity'][i] + ' ' + data['Alcohol'][i] + ' ' + data['Frightening'][i])
        
    return important_features
```

```python
dataFrame['important_features'] = get_important_features(dataFrame)
dataFrame.head(3)
```

```python
dataClean = pd.DataFrame(dataFrame.important_features.apply(text_temizle1))
dataClean
```

```python
cm = CountVectorizer().fit_transform(dataFrame['important_features'])
```

```python
cs = cosine_similarity(cm)
print(cs)
```

```python
cs.shape
```

```python
title = 'The Human Centipede First Sequence'

movie_id = dataFrame[dataFrame.Name == title]['Unnamed: 0'].values[0]
```

```python
movie_id
```

```python
scores = list(enumerate(cs[movie_id]))
```

```python
sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)
sorted_scores = sorted_scores[1:]
```

```python
print(sorted_scores)
```

```python
j = 0
print('The 10 most recomended movies to', title, 'are:\n')
for item in sorted_scores:
    movieTitle = dataFrame[dataFrame['Unnamed: 0'] == item[0]]['Name'].values[0]
    print(j+1, movieTitle)
    j+=1
    if j > 9:
        break
```
